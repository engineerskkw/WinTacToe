### Imports

```python
from reinforcement_learning import *
from environments import TicTacToeEngine
```

### Engine initialization

```python
engine = TicTacToeEngine(2, 3, 3)
```

### Agents initialization

```python
# agents = [QLearningAgent(0.3,0.1,0.9),
#           BasicAgent(0.1)]
random_agent = RandomAgent()
first_player = QLearningAgent(0.3, 0.1, 0.9)
second_player = QLearningAgent(0.3, 0.1, 0.9)
```

### Training

```python
# Pre-training with non-learning random agent

with SimpleTraining(engine, [first_player, random_agent]) as st:
    agents = st.train(50000)
first_player = agents[0]

with SimpleTraining(engine, [random_agent, second_player]) as st:
    agents = st.train(50000)
second_player = agents[1]
```

```python
first_player.visualize()
second_player.visualize()
```

```python
# Training with both learning agents
with SimpleTraining(engine, [first_player, second_player]) as st:
    agents = st.train(1000)
    
first_player = agents[0]
second_player = agents[1]
```

```python
first_player.visualize()
second_player.visualize()
```

```python
# Training with both learning agents
with SimpleTraining(engine, [first_player, second_player]) as st:
    agents = st.train(10000)
    
first_player = agents[0]
second_player = agents[1]
```

### Visualization

```python
first_player.visualize()
second_player.visualize()
```

```python
with SimpleTraining(engine, [first_player, second_player]) as st:
    agents = st.train(50000)
    
first_player = agents[0]
second_player = agents[1]
```

```python
first_player.visualize()
second_player.visualize()
```

### Epsilon 0 phase

```python
first_player.epsilon = 0.0
second_player.epsilon = 0.0
```

```python
with SimpleTraining(engine, [first_player, second_player]) as st:
    agents = st.train(10000)
    
first_player = agents[0]
second_player = agents[1]
```

```python
first_player.visualize()
second_player.visualize()
```
